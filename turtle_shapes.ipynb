```python
import turtle as t

# set the window title
t.title("Shapes")

# start with a blank window
t.reset()

# set the speed of the turtle (1 - slowest, 10 - fast, 0 - no animation)
t.speed(6)

# set the shape of the turtle
t.shape("turtle")

# set the background color
t.bgcolor("light green")

# draw a triangle
t.pendown()
t.forward(150)   # move forward 150 pixels
t.left(120)      # turn left 120 degrees
t.forward(150)
t.left(120)
t.forward(150)

# move to a new location
t.penup()
t.goto(-100, -200)
t.setheading(180)   # set turtle to face left

# draw an outlined rectangle 
t.pendown()
t.pensize(10)
t.color("blue", "white")  # color(pen_color, fill_color)
t.begin_fill()
t.forward(200)
t.right(90)
t.forward(100)
t.right(90)
t.forward(200)
t.right(90)
t.forward(100)
t.right(90)
t.end_fill()

# move to a new location
t.penup()
t.goto(-200, 150)

# draw a star-like object
# a better implementation would use a loop instead
#    of copy and paste.
t.setheading(90)   # face upward
t.pendown()
t.color("yellow")
t.forward(100)
t.backward(100)
t.right( 360/5 )   # turn 360/5 degrees so I have 5 equal arms
t.forward(100)
t.backward(100)
t.right( 360/5 )
t.forward(100)
t.backward(100)
t.right( 360/5 )
t.forward(100)
t.backward(100)
t.right( 360/5 )
t.forward(100)
t.backward(100)
t.right( 360/5 )

# move to another location
t.penup()
t.goto(150, -300)
t.setheading( 0 )  # face to the right
t.color("red")
t.pensize(2)

# use the circle command to draw regular polygons
# a regular polygon has all sides and angles equal
t.pendown()
#t.circle(150)
t.circle(150, 360, 5)  # circle(radius, extent, number of points)
                       # extent is the number of degrees around to draw
                       # 3 points will draw a triangle
                       # 4 points will draw a square, etc.

t.hideturtle()
```
